```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv(r"D:\data_projects\cancer_data\china_cancer_patients_synthetic.csv")
```

```python
df['DiagnosisDate']=pd.to_datetime(df['DiagnosisDate'])
```

```python
df['SurgeryDate']=pd.to_datetime(df['SurgeryDate'])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
categorical_columns=['AlcoholUse','GeneticMutation','Comorbidities']
for col in categorical_columns:
    df[col]=df[col].mode()[0]
df['SurgeryDate']=df['SurgeryDate'].fillna(df['DiagnosisDate'])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum() 
```

```python
columns=['Age','TumorSize','ChemotherapySessions','RadiationSessions','FollowUpMonths']
for col in columns:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    outliers=((df[col]<lower_bound) | (df[col]>upper_bound))
    print(f"{col} has {outliers} outliers")
```

```python
categorical_columns=['Gender','Province','Ethnicity','TumorType','CancerStage','Metastasis','TreatmentType','SurvivalStatus'
,'SmokingStatus','AlcoholUse','GeneticMutation','Comorbidities']
for col in categorical_columns:
    print(f"{col} has {df[col].value_counts()} values" )
```

```python
# a new feature
# 1: add age group
labels=['<30','30-50','50-70','70+']
bins=[0,30,50,70,120]
df['AgeGroup']=pd.cut(df['Age'],labels=labels,bins=bins)
# 2: add tumor category
labels=['Small','Medium','Large']
bins=[0,2,5,100]
df['TumorCategory']=pd.cut(df['TumorSize'],labels=labels,bins=bins)
```

```python
df.head()
```

```python
categorical_columns=['Gender','Province','Ethnicity','TumorType','CancerStage','Metastasis','TreatmentType','SurvivalStatus'
,'SmokingStatus','AlcoholUse','GeneticMutation','Comorbidities']
for col in categorical_columns:
    print(f"{col} : {df[col].describe()}")
```

Analysis of the Distribution of Types and Size of Tumors by Gender

```python
plt.figure(figsize=(15,9))
sns.boxplot(data=df,x='TumorType',y='TumorSize',hue='Gender',palette='Pastel1')
plt.title('Tumor Size Distribution by Tumor Type and Gender')
plt.xlabel('Tumor Type')
plt.ylabel('Tumor Size (in cm)')
plt.tight_layout()
plt.show()
```

Analysis of Cancer Stages by Age Group and Gender

```python
Gender = sns.FacetGrid(df, col='Gender', height=6, aspect=1.2)
age_order=['<30','30-50','50-70','70+']
Gender.map(sns.countplot,'AgeGroup', hue='CancerStage',data=df, palette='Pastel1',order=age_order,hue_order=['I', 'II', 'III', 'IV'])
Gender.fig.suptitle('Cancer Stage Distribution by Age Group for each Gender')
Gender.add_legend(title='Cancer Stage')
Gender.fig.suptitle('Cancer Stage Distribution by Age Group for each Gender')
Gender.add_legend(title='Cancer Stage',bbox_to_anchor=(1.05,1),loc='upper left')
for ax in Gender.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(90)
plt.tight_layout()
plt.show()
```

Correlation Matrix

```python
numerical_df=df.select_dtypes(['int64','float64'])
corr_matrix=numerical_df.corr()
plt.figure(figsize=(8, 10))
sns.heatmap(data=corr_matrix,annot=True,fmt='.2f',cmap='Blues')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

Analysis of the Distribution of Types of Tumors by Age Group

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='AgeGroup',hue='TumorType',palette='Purples')
plt.title('Distribution of Types of Tumors by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.show()
```

Analysis of the Preferred Remedy for each Age Group and Gender

```python
Gender = sns.FacetGrid(df, col='Gender', height=6, aspect=1.2)
age_order=['<30','30-50','50-70','70+']
Gender.map(sns.countplot,'AgeGroup', hue='TreatmentType',data=df, palette='PRGn',order=age_order)
Gender.fig.suptitle('Treatment Type Distribution by Age Group for each Gender')
Gender.fig.suptitle('Cancer Stage Distribution by Age Group and Gender')
Gender.add_legend(title='Treatment Type',bbox_to_anchor=(1.05,1),loc='upper left')
for ax in Gender.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(45)
plt.tight_layout()
plt.show()
```

Analysis of Somking Status and its Relationship to Tumor Size by Gender

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df,x='SmokingStatus',y='TumorSize',hue='Gender',palette='PRGn')
plt.title('Tumor Type Distribution by Smoking Status and Gender')
plt.xlabel('Smoking Status')
plt.ylabel('Tumor Size (in cm)')
plt.tight_layout()
plt.show()
```

The Effect of Tumor Type on Survival Status

```python
survival_by_gender=pd.crosstab(df['TumorType'],df['SurvivalStatus'],normalize='index')*100
survival_by_gender.plot(kind='bar',stacked=True)
sns.set_palette('Pastel2')
plt.title('Survival Status Tumor Type (%)')
plt.xlabel('Gender')
plt.ylabel('Percentage of Patients')
Gender.add_legend(title='Survival Status')
Gender.add_legend(title='Cancer Stage',bbox_to_anchor=(1.05,1),loc='upper left')
plt.tight_layout()
plt.show()
```
